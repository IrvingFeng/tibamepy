## 比特幣分析

### 抓取資料

```python
import requests
res = requests.get('https://www.coingecko.com/chart/1/usd.json?locale=en')
stats = res.json()['stats']
```

### 資料整理

```python
import pandas
df = pandas.DataFrame(stats)
```

```python
df.columns = ['datetime', 'btc']
```

```python
df.set_index('datetime', inplace=True)
```

### 視覺化

```python
%pylab inline
df.plot(kind='line')
```

### 產生均線

```python
df['mvg30'] = df['btc'].rolling(window = 30).mean()
```

```python
%pylab inline
df.plot(kind='line')
```
